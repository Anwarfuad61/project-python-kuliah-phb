```python
import tkinter as tk
from tkinter import filedialog, messagebox
import cv2
import numpy as np
from tkinter import font as tkfont
from tkinter import messagebox,PhotoImage
from PIL import ImageTk, Image
import os
import joblib
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from sklearn.neighbors import KNeighborsClassifier
#from tkinter.scrolledtext import ScrolledText

class MainUI(tk.Tk):
    def __init__(self):
        tk.Tk.__init__(self)
        self.title_font = tk.font.Font(family='Helvetica', size=18, weight="bold")
        self.geometry("600x400")  # Set your desired window size
        self.title("Deteksi Kematangan Buah")
        
        container = tk.Frame(self)
        container.pack(side="top", fill="both", expand=True)
        container.grid_rowconfigure(0, weight=1)
        container.grid_columnconfigure(0, weight=1)

        self.frames = {}
        for F in (halawal,haltrain,haldeteksi,halpredik):
            page_name= F.__name__
            frame = F(parent=container, controller=self)
            self.frames[page_name]= frame
            frame.grid(row=0, column=0, sticky="nsew")

        self.show_frame("halawal")

    def show_frame(self, page_name):
            frame = self.frames[page_name]
            frame.tkraise()

            
class halawal(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        
        label = tk.Label(self, text="Home Page", font=self.controller.title_font, fg="#263942")
        label.grid(row=0, sticky="ew", padx=10, pady=10)

        button1 = tk.Button(self, text="Create Klasifikasi", fg="#ffffff", bg="#263942", padx=30, pady=5,
                            command=lambda: self.controller.show_frame("haltrain"))
        button2 = tk.Button(self, text="Predict Otomatis", fg="#ffffff", bg="#263942", padx=30, pady=5,
                            command=lambda: self.controller.show_frame("haldeteksi"))
        button4 = tk.Button(self, text="Predict Manual", fg="#ffffff", bg="#263942", padx=30, pady=5,
                            command=lambda: self.controller.show_frame("halpredik"))
        button3 = tk.Button(self, text="Quit", fg="#263942", bg="#ffffff", padx=20, pady=5, 
                            command=self.on_closing)

        button1.grid(row=1, column=3, pady=10)
        button2.grid(row=2, column=3, pady=10)
        button3.grid(row=4, column=3, pady=10)
        button4.grid(row=3, column=3, pady=10)


    def on_closing(self):
        if messagebox.askokcancel("Quit", "Are you sure?"):
            self.controller.destroy() 

class haltrain(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        self.lbl_folder_path = tk.Label(self, text="Folder Path", fg="#263942", font='Helvetica 12 bold')
        self.lbl_folder_path.grid(row=0, column=0, pady=10, padx=5)
        self.path_entry = tk.Entry(self, borderwidth=3, bg="lightgrey", font='Helvetica 11')
        self.path_entry.grid(row=0, column=1, pady=10, padx=10)
        
        tk.Label(self, text="Nama Model", fg="#263942", font='Helvetica 12 bold').grid(row=1, column=0, padx=10, pady=10)
        self.model_name = tk.Entry(self, borderwidth=3, bg="lightgrey", font='Helvetica 11')
        self.model_name.grid(row=1, column=1, pady=10, padx=10)
        
        tk.Label(self, text="Jumlah Epoch", fg="#263942", font='Helvetica 12 bold').grid(row=2, column=0, padx=10, pady=10)
        self.epoch_entry = tk.Spinbox(self, from_=1, to=100, font='Helvetica 11')
        self.epoch_entry.grid(row=2, column=1, pady=10, padx=10)
        
        self.buttonselc = tk.Button(self, text="Select Folder", fg="#ffffff", bg="#263942",command=self.selectfol)
        self.buttonselc.grid(row=3, column=0, pady=10, ipadx=5, ipady=4)

        self.buttoncanc = tk.Button(self, text="Kembali", bg="#ffffff", fg="#263942",command=lambda: controller.show_frame("halawal"))
        self.buttoncanc.grid(row=4, column=1, pady=10, ipadx=5, ipady=4)

        self.buttonext = tk.Button(self, text="train", fg="#ffffff", bg="#263942",command=self.start_training)
        self.buttonext.grid(row=3, column=1, pady=10, ipadx=5, ipady=4)
        
        self.buttoncanc = tk.Button(self, text="Use Default Model", bg="#45aaf2", fg="#263942",command=self.start_default)
        self.buttoncanc.grid(row=3, column=2, pady=10, ipadx=5, ipady=4)
    
    def selectfol(self):
        try:
            self.folder_path = filedialog.askdirectory(initialdir="/images", title="Select Folder")
            if self.folder_path:
                self.path_entry.delete(0, tk.END)
                self.path_entry.insert(0, self.folder_path)
        except Exception as e:
            tk.messagebox.showerror("Error", str(e))
    
    #melakukan train     
    def start_training(self):
        if hasattr(self, 'folder_path'):
        # Mendapatkan nilai jumlah epoch dari entry
            epoch_input = self.epoch_entry.get()
            if epoch_input.isdigit():  # Memeriksa apakah input adalah digit
                num_epochs = int(epoch_input)

                # Melakukan proses pelatihan hanya jika nilai epoch valid
                train_datagen = ImageDataGenerator(rescale=1./255)

                train_generator = train_datagen.flow_from_directory(
                    self.folder_path,
                    target_size=(128, 128),
                    batch_size=32,
                    class_mode='categorical')

                # Ubah model menjadi KNN dengan n_neighbors yang berbeda
                knn_model = KNeighborsClassifier(n_neighbors=5, weights='distance', metric='euclidean')
                
                # Dapatkan data dan label dari generator
                train_images, train_labels = [], []
                for _ in range(len(train_generator)):
                    batch_images, batch_labels = train_generator.next()
                    train_images.extend(batch_images)  # Kumpulkan gambar-gambar dari generator
                    train_labels.extend(batch_labels)

                # Dapatkan data dan label dari generator
                train_images, train_labels = train_generator.next()
                train_images = train_images.reshape(train_images.shape[0], -1)

                # Melatih model KNN
                knn_model.fit(train_images, np.argmax(train_labels, axis=1))

                # Mendapatkan nama model dari Entry
                model_name = self.model_name.get()

                # Simpan model ke dalam folder_latih setelah melatih
                folder_latih = 'folder_latih'
                if not os.path.exists(folder_latih):
                    os.makedirs(folder_latih)

                model_path = os.path.join(folder_latih, f'{model_name}_model.sav')  # Gunakan nama model dari input
                joblib.dump(knn_model, model_path)  # Simpan model menggunakan joblib

                tk.messagebox.showinfo("Training Completed", f"Model '{model_name}' has been trained and saved successfully!")
            else:
                tk.messagebox.showerror("Invalid Input", "Please enter a valid number for Epochs.")
                
    def start_default(self):
        default_model_path = 'latih_default/trained_model.h5'  # Path ke model default

        if os.path.exists(default_model_path):
            # Lakukan sesuatu dengan model default
            tk.messagebox.showinfo("Success", "Using default model!")
            self.default_model_path = default_model_path  # Menyimpan path model default
        else:
            tk.messagebox.showerror("Error", "Default model not found.")

class haldeteksi(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent, bg='#45aaf2')
        self.controller = controller
        self.filename = None
        self.img = None

        self.lbl_pic_path = tk.Label(self, text='Image Path:', fg="#263942", font='Helvetica 12 bold')
        self.lbl_pic_path.grid(row=0, column=0, pady=10, padx=5)

        self.lbl_show_pic = tk.Label(self, bg='#45aaf2')
        self.lbl_show_pic.grid(row=1, column=0, columnspan=2)

        self.entry_pic_path = tk.Entry(self, borderwidth=3, bg="lightgrey", font='Helvetica 11')
        self.entry_pic_path.grid(row=0, column=1, pady=10, padx=10)

        self.btn_browse = tk.Button(self, text='Select Image',fg="#263942", font='Helvetica 12 bold', command=self.selectPic)
        self.btn_browse.grid(row=2, column=1,pady=10, ipadx=5, ipady=4)
        
        btn_process = tk.Button(self, text='Process default',bg="#568203", fg="#FFFFFF", font='Helvetica 11 bold')
        btn_process ['command']= self.process_image  # Menghubungkan tombol dengan fungsi process_image
        btn_process.grid(row=2, column=2, pady=10, ipadx=5, ipady=4)  # Menambahkan grid untuk tombol
        
        self.result_label = tk.Label(self, text="", fg="#263942", font='Helvetica 12 bold')
        self.result_label.grid(row=4, column=0, pady=10, ipadx=5, ipady=4)  # Menambahkan grid untuk label hasil

        self.buttoncanc = tk.Button(self, text="Cancel", bg="#800000", fg="#F6F4F1",command=lambda: controller.show_frame("halawal"))
        self.buttoncanc.grid(row=0, column=4,  pady=10, ipadx=5, ipady=4)    

    def selectPic(self):
        try:
            self.filename = filedialog.askopenfilename(
                initialdir="/images",
                title="Select Image",
                filetypes=(("png images", "*.png"), ("jpg images", "*.jpg"))
            )
            if self.filename:
                self.img = Image.open(self.filename)
                self.img = self.img.resize((200, 200), Image.ANTIALIAS)
                self.img = ImageTk.PhotoImage(self.img)
                self.lbl_show_pic['image'] = self.img
                self.entry_pic_path.delete(0, tk.END)
                self.entry_pic_path.insert(0, self.filename)

        except Exception as e:
            tk.messagebox.showerror("Error", str(e))

    def process_image(self):
        if self.filename:
            try:
                default_model_path = 'latih_default/trained_model.h5'
                if os.path.exists(default_model_path):
                    result = self.process_image_with_model(self.filename, default_model_path)
                    self.result_label.config(text=result)
                    self.result_label.grid(row=3, column=0, columnspan=2)
                else:
                    tk.messagebox.showerror("Error", "Default model not found.")
            except Exception as e:
                tk.messagebox.showerror("Error", str(e))
        else:
            tk.messagebox.showerror("Error", "Please select an image.")

    def process_image_with_model(self, filename, default_model_path):
        try:
            # Load model yang telah dilatih
            model = tf.keras.models.load_model(default_model_path)

            # Ubah gambar menjadi format yang dapat diterima oleh model
            img = Image.open(filename)
            img = img.resize((128, 128))  # Sesuaikan dengan ukuran yang diharapkan oleh model Anda
            img = tf.keras.preprocessing.image.img_to_array(img)
            img = tf.expand_dims(img, axis=0)

            # Prediksi menggunakan model
            prediction = model.predict(img)
            predicted_class = tf.argmax(prediction, axis=1).numpy()[0]

            # Kembalikan hasil berdasarkan prediksi kelas
            classes = ['mentah', 'setengah matang', 'matang']  # Sesuaikan dengan kelas yang ada di model Anda
            result = f"Prediksi Kematangan Buah: {classes[predicted_class]}"
            return result

        except Exception as e:
            return f"Error processing image: {str(e)}"
        
class halpredik(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        self.filename = None
        self.img = None

        self.lbl_pic_path = tk.Label(self, text='Image Path:', fg="#263942", font='Helvetica 12 bold')
        self.lbl_pic_path.grid(row=0, column=0, pady=10, padx=5)

        self.lbl_show_pic = tk.Label(self, bg='#45aaf2')
        self.lbl_show_pic.grid(row=1, column=0, columnspan=2)

        self.entry_pic_path = tk.Entry(self, borderwidth=3, bg="lightgrey", font='Helvetica 11')
        self.entry_pic_path.grid(row=0, column=1, pady=10, padx=10)

        self.btn_browse = tk.Button(self, text='Select Image',fg="#263942", font='Helvetica 12 bold', command=self.selectPic)
        self.btn_browse.grid(row=2, column=1,pady=10, ipadx=5, ipady=4)
        
        self.result_label = tk.Label(self, text="", fg="#263942", font='Helvetica 12 bold')
        self.result_label.grid(row=4, column=0, pady=10, ipadx=5, ipady=4)  # Menambahkan grid untuk label hasil

        self.buttoncanc = tk.Button(self, text="Cancel", bg="#800000", fg="#F6F4F1",command=lambda: controller.show_frame("halawal"))
        self.buttoncanc.grid(row=0, column=4,  pady=10, ipadx=5, ipady=4)
        
        self.label_model = tk.Label(self, text="Model Path:", fg="#263942", font='Helvetica 12 bold')
        self.label_model.grid(row=1, column=3, padx=10, pady=10)

        self.model_path = tk.StringVar()
        self.model_entry = tk.Entry(self, textvariable=self.model_path, font='Helvetica 11', state='disabled')
        self.model_entry.grid(row=1, column=4, pady=10, padx=10)

        self.browse_button = tk.Button(self, text="Browse", command=self.browse_model)
        self.browse_button.grid(row=2, column=4, pady=10, padx=10)
        
        self.process_button1 = tk.Button(self, text="Process pilih",bg="#568203", fg="#FFFFFF", font='Helvetica 11 bold', command=self.process_images)
        self.process_button1.grid(row=3, column=4, pady=10, padx=10)
        
    def process_images(self):
        model_path = self.model_path.get()
        if not model_path:
            tk.messagebox.showerror("Error", "Please select a model file first.")
            return
        if hasattr(self, 'filename'):
            # Load model, dapatkan gambar, dan ubah format gambar
            knn_model = joblib.load(model_path)

            img = Image.open(self.filename)
            img = img.resize((128, 128))  # Sesuaikan dengan ukuran yang diharapkan oleh model Anda
            img = np.array(img)
            img = img.reshape(1, -1)

            # Prediksi menggunakan model KNN
            predicted_class = knn_model.predict(img)

            ## Kembalikan hasil berdasarkan prediksi kelas
            classes = ['mentah', 'setengah matang', 'matang']  # Sesuaikan dengan kelas yang ada di model Anda
            result = f"Prediksi Kematangan Buah: {classes[int(predicted_class)]}"
            tk.messagebox.showinfo("Prediction", result)
            return result

    def browse_model(self):
        filename = filedialog.askopenfilename(initialdir="/folder_latih", title="Select Model File",
                                              filetypes=(("SAV files", "*.sav"), ("All files", "*.*")))
        if filename:
            self.model_path.set(filename)  

    def selectPic(self):
        try:
            self.filename = filedialog.askopenfilename(
                initialdir="/images",
                title="Select Image",
                filetypes=(("png images", "*.png"), ("jpg images", "*.jpg"))
            )
            if self.filename:
                self.img = Image.open(self.filename)
                self.img = self.img.resize((200, 200), Image.ANTIALIAS)
                self.img = ImageTk.PhotoImage(self.img)
                self.lbl_show_pic['image'] = self.img
                self.entry_pic_path.delete(0, tk.END)
                self.entry_pic_path.insert(0, self.filename)

        except Exception as e:
            tk.messagebox.showerror("Error", str(e))
            
    
        
app = MainUI()
app.mainloop()
```
